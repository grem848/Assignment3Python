```python
plt.close()
# 7: Create illustration 2: In your notebook use your module to create a dataset where development of all German citizens can be shown over time

# first creating the dataset with only Germans included
german_mask = bef_stat[:,3] == 5180
german_data = module.mask_function(bef_stat, german_mask)

# then running it through dict function and printing the graph
german_dict = module.get_dict_of_x_values_and_corresponding_sums(german_data, 'year')
title = 'German citizens over time'
plt.title(title, fontsize=12)
plt.plot(german_dict.keys(), german_dict.values())
plt.show()
```

```python
# Create illustration 3: Show a bar plot of all the 18-25 year old in the different areas of copenhagen and in the same plot 
# show (in a different color) all 60-67 year old in the different areas.
plt.close()

young_people = module.get_age_group_per_year(bef_stat, 18, 25, 2015)

old_people = module.get_age_group_per_year(bef_stat, 60, 67, 2015)

young_dict = module.get_dict_of_x_values_and_corresponding_sums(young_people, 'area')
old_dict = module.get_dict_of_x_values_and_corresponding_sums(old_people, 'area')

young_values = list(young_dict.values())
old_values = list(old_dict.values())

area_codes = np.unique(bef_stat[:,1])
index = np.arange(1, len(area_codes)+1)
plt.bar(index, young_values, width=0.5, linewidth=0, align='center')
title = '18-25 yearolds vs. 60-67 yearolds over neighbourhoods'
plt.title(title, fontsize=12)
plt.xlabel("Neighbourhoodcode", fontsize=10)
plt.ylabel("Amount", fontsize=10)
plt.xticks(index, area_codes, fontsize=10)
plt.bar(index, old_values, width=0.5, linewidth=0, align='center',color='green', bottom=young_values)
plt.show()
```

```python
# Create illustration 4: show a pie chart of age groups (0-10, 10-20, 20-30 ...) in Ã˜sterbro part
#  of Copenhagen. create another similar piechart for Vesterbro.
plt.close()

# Ã¸sterbro = 2
year = 2015
list_of_age_group_sums = module.get_list_of_age_group_sums(2, year, bef_stat)

age_groups = ['0-10', '11-21', '21-30', '31-40', '41-50', '51-60', '61-70', '71-80', '81-90', 'Above 90']

labels = age_groups
sizes = list_of_age_group_sums
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')
title = 'Age distribution Ã˜sterbro ' + str(year)
plt.title(title, fontsize=12)
plt.show()
```
